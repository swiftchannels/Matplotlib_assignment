```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
Trial_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
combined_data = pd.merge(mouse_data,Trial_data,how='right',on='Mouse ID')

# Display the data table for preview

combined_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
grouped_drug_Timepoint = combined_data.groupby(['Drug', 'Timepoint']).mean()['Tumor Volume (mm3)']
grouped_drug_Timepoint = grouped_drug_Timepoint.reset_index()
# Convert to DataFrame
Grouped_data_df = pd.DataFrame(grouped_drug_Timepoint)
# Preview DataFrame
Grouped_data_df.head()
```

```python
grouped_sem = combined_data.groupby(['Drug', 'Timepoint']).sem()['Tumor Volume (mm3)']
grouped_sem = grouped_sem.reset_index()
grouped_sem_df = pd.DataFrame(grouped_sem)
grouped_sem.head()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
Tumor_Volume_initial = munging_data.loc[0,:]
Tumor_Volume_final = munging_data.loc[45,:]
percent_change = ((Tumor_Volume_final - Tumor_Volume_initial)/Tumor_Volume_initial)*100
# Display the data to confirm
percent_change.head()
```

```python
# Store all Relevant Percent Changes into a Tuple
percent_change_tuple = list(zip(percent_change.index,percent_change))
percent_change_list = list(percent_change)
passing = []
failing = []
for value in percent_change_list:
    if value > 0:
        failing.append(value)
    else:
        passing.append(value)
#plot location
plt.figure(figsize=(12,6))
# defining y axis and x axis
drugs_plot = [percent_change['Capomulin'],percent_change['Infubinol'],percent_change['Ketapril'],percent_change['Placebo']]

x_axis = np.arange(len(drugs_plot))
color = ['g','r','r','r']

drugsplot = plt.bar(x_axis, drugs_plot , color=color, alpha=1, align = "edge", width = -1)
#set plot properties
x_labels = ["Capomulin", "Infubinol", "Ketapril", "Placebo"]
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, x_labels)
plt.ylim(-30, max(drugs_plot)+10)
plt.yticks=[-20, 0, 20, 40, 60]
plt.ylabel('% Tumor Volume Change')
plt.title('Tumor Change Over 45 Day Treatment')
plt.grid()
plt.tight_layout()
# Use functions to label the percentages of changes
def autolabel(plot):
    for bar in plot:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, .1*height, "%d" %int(height)+ "%",
                ha='center', va='top', color="white")
    
# Call functions to implement the function calls
autolabel(drugsplot)
# safe graph
plt.savefig("Tumor_change.png")
plt.show()
```
